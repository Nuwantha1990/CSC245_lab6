```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv("company_sales_data.csv")
monthList = df['month_number'].tolist()
monthly_facecrem= df['facecream'].tolist()
monthly_facewash= df['facewash'].tolist()
monthly_toothpaste= df['toothpaste'].tolist()
monthly_bathingsoap= df['bathingsoap'].tolist()
monthly_shampoo= df['shampoo'].tolist()
monthly_moisturizer= df['moisturizer'].tolist()

plt.plot(monthList, monthly_facecrem, marker='o', linestyle='-', color='red',markersize=8,linewidth=3,label='Face Cream Sales Data')
plt.plot(monthList, monthly_facewash, marker='o', linestyle='-', color='green',markersize=8,linewidth=3,label='Face Wash Sales Data')
plt.plot(monthList, monthly_toothpaste, marker='o', linestyle='-', color='blue',markersize=8,linewidth=3,label='Toothpaste Sales Data')

plt.plot(monthList, monthly_bathingsoap , marker='o', linestyle='-', color='yellow',markersize=8,linewidth=3,label='Bathing Soap Sales Data')

plt.plot(monthList, monthly_shampoo , marker='o', linestyle='-', color='purple',markersize=8,linewidth=3,label='Shampoo Sales Data')

plt.plot(monthList, monthly_moisturizer , marker='o', linestyle='-', color='orange',markersize=8,linewidth=3,label='Moisturizer Sales Data')
plt.grid(True,linewidth='1',linestyle='-')
plt.ylabel('Sales Units in Numbers', fontsize=12, fontweight='bold')
plt.xlabel('Month Number', fontsize=12, fontweight='bold')
plt.title('Sales Data')
plt.legend(loc='upper left',fontsize=10)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('company_sales_data.csv')
monthList = df['month_number'].tolist()
monthly_facewash= df['facewash'].tolist()
monthly_facecream = df['facecream'].tolist()
# Setting width of each bar
barWidth = 0.3

# Set position of bar on X axis
r1 = np.arange(len(monthly_facewash))
r2 = [x + barWidth for x in r1]

# Create the bars for facewash sales
plt.bar(r1, monthly_facecream, color='blue', width=barWidth, label='Facewash Sales')

# Create the bars for facecream sales
plt.bar(r2, monthly_facewash, color='orange', width=barWidth, label='Facecream Sales')

# Add xticks on the middle of the group bars
plt.xlabel('Month Number', fontsize=12)
plt.ylabel('Sales Units in Numbers', fontsize=12, fontweight='bold')
plt.xticks([r + barWidth/2 for r in range(len(monthly_facewash))], monthList)

plt.title('Facewash and Facecream Sales Data')
plt.legend(loc='upper left', fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('company_sales_data.csv')
#calculate tottal sale  data for last year for each product
total_sales_per_product = df.iloc[:, 1:].sum(axis=0)
#ploting the pie chart
plt.figure(figsize=(8,8))
plt.pie(total_sales_per_product,labels=df.columns[1:], autopct='%1.1f%%',startangle=140)
plt.title('Number of Units Sold per Year for Each Product (Last Year)')
plt.axis('equal')
plt.legend(loc='lower right')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df =pd.read_csv('company_sales_data.csv')
# Set up subplots
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# Plot Bathing soap data
axes[0].plot(df['month_number'], df['bathingsoap'], marker='o', color='b')
axes[0].set_title('Bathing Soap Sales Over Months')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Units Sold')

# Plot Facewash data
axes[1].plot(df['month_number'], df['facewash'], marker='o', color='r')
axes[1].set_title('Facewash Sales Over Months')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Units Sold')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
